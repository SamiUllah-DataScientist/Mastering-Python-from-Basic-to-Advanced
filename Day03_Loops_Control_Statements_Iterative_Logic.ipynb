```python
# reversing Number
digt = int(input("Enter a 5-digit Number: "))
n = 1
while n <= digt:
    num = digt % 10
    digt = digt // 10
    print(num)
```

```python
# ATM Simulation

acc_holder = input("Enter Account Holder Name: ")
acc_pin = 1234
balance = 100000

# ✅ FIX 1: PIN check karo - 3 chances diye hain
attempts = 0
while attempts < 3:
    correct_pin = int(input("Enter the PIN Code: "))
    if correct_pin == acc_pin:
        print("✅ PIN Correct! Welcome,", acc_holder)
        break
    else:
        attempts += 1
        print(f"❌ Wrong PIN! {3 - attempts} attempts remaining.")

if attempts == 3:
    print("🔒 Account Blocked! Too many wrong attempts.")
    exit()  # PIN galat tha toh program band

# ✅ Main ATM Loop
while True:
    print("=" * 30)
    print("      ATM Simulation     ")
    print("=" * 30)
    print("1. Check Balance")
    print("2. Deposit Money")       # ✅ FIX 3: Spelling fix
    print("3. Withdraw Money")
    print("4. Show Dashboard")
    print("5. Exit")

    choice = int(input("Enter a Choice: "))

    if choice == 1:
        print("=" * 30)
        print(f"  Account Balance     | Rs.{balance}")
        print("-" * 30)

    elif choice == 2:
        amount = int(input("Enter the Deposit Amount: "))
        if amount <= 0:
            print("  Invalid amount!")
        else:
            balance += amount
            print("  ✅ Money Deposited Successfully!")
            print("=" * 30)
            print(f"  Deposited Amount    | Rs.{amount}")
            print(f"  Account Balance     | Rs.{balance}")
            print("-" * 30)

    elif choice == 3:
        withdraw_amount = int(input("Enter the Withdrawal Amount: "))
        if withdraw_amount > balance:
            print("  ❌ Sorry! Insufficient Balance.")
        elif withdraw_amount < 500:
            print("  ❌ Sorry! Minimum withdrawal is Rs.500")
        else:
            balance -= withdraw_amount
            print("  ✅ Money Withdrawn Successfully!")  # ✅ FIX 4: Message fix
            print("=" * 30)
            print(f"  Withdrawal Amount   | Rs.{withdraw_amount}")
            print(f"  Account Balance     | Rs.{balance}")
            print("-" * 30)

    elif choice == 4:
        print("-------  Account Dashboard  -------")
        print(f"  Account Holder    | {acc_holder}")
        print(f"  Account Balance   | Rs.{balance}")
        print("-" * 35)

    elif choice == 5:
        print("  Allah Hafiz! Thank You 😊")
        break

    else:
        print("  ❌ Invalid Choice! Try Again.")
```
