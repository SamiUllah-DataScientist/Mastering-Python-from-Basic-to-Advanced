## 🚀 Day 01: The Python Foundation Challenge

### Task 1: Memory Management & Object Identity

```python
x = 500 
y = 500
print(id(x))
print(id(y))
print("Memory ID Comparison:", id(x) == id(y))
x = y 
print("After Equal Memory ID Comparison:", id(x) == id(y))
```

### Task 2: Dynamic Salary Calculator (Type Casting)

```python
salary = input("Enter the Salary: ")
allowance_percentage = input("Enter the Allowance Percentage: ")
tax_det_percentage = input("Enter the Tax Detection Percentage: ") 
# direct convert the type and calculate the total bill
final_amount = float(salary) + int(allowance_percentage) + int(tax_det_percentage)
print("Take-Home Salary:", int(final_amount))
```
