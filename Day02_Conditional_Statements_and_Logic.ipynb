```python
num = int(input("Enter a Number: "))
""" this is a if else block condition if condition met then if block runs,
if condtion does not met , then else block runs."""
if num % 2 == 0:
    print(f"{num} is a Even Number.")
else:
    print(f"{num} is a Odd Number.")
```

```python
marks = int(input("Enter Marks(1-100): "))
"""this is a if elif else condition complier checks form top to bottom 
if complier reads any correct statements then if runs that condition otherwise runs else block"""
if 0 < marks <= 100:
    
    if 90 <= marks <= 100:
        print("Grade: A+")
    elif 80 <= marks < 90:
        print("Grade: A")
    elif 70 <= marks< 80:
        print("Grade: B")
    elif 60 <= marks < 70:
        print("Grade: C")
    elif 50 <= marks < 60:
        print("Grade: D")
    elif 40 <= marks < 50:
        print("Grade: E")
    else:
        print("Grade: F")
else:
    print("Invalid Marks. Input out of the valid range...")
```

```python
marks = 44
# this is a short handed form condition mean it is written in only one single line 
status = f"Marks is: {marks} which is {'Pass' if marks >= 60 else 'Fail'}"
print(status)
```

## **Grand Task: ATM Simulation**

```python
balance = int(input("Enter the Account Balance: "))
withdraw_amount = int(input("Enter the Withdraw Amount: "))
print("=" * 30)
print("       ATM Simulation        ")
print("-" * 30)

if withdraw_amount <= 0:
    print("Invalid Amount! Please enter a positive value.")
elif withdraw_amount > balance:
    print("Insufficient Funds")
else:
    # Transaction Success Logic
    if withdraw_amount < 500:
        print("Minimum withdrawal is 500")
    else:
        balance -= withdraw_amount
        print(f"Congratulations! You are successful in withdraw Money.\nRemaining Balance is {balance}.")
```
