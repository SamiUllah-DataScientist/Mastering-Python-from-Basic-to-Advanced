# **Tuple**
A tuple in python is sismilar to a list. the difference between the two is taht we cannot change the elements of a tupleonce it is created assigned whereas we can change the elements of a list

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.
## Charatertistics
* Ordered
* Unchanged
* Allows Duplicates

```python
# empty tuple 
t1 = ()
print(t1)
# single item tuple 
# t2 = (2)
t2 = (2,)
print(t2)
# homo tuple
t3 = (1,2,3,4)
print(t3)
# heterogeneous
t4 = (1,2,5, True, [7,8,6])
print(t4)
# 2D tuple
t5 = (1,2,3,(5,6))
print(t5)
# using type conversion
t6 = tuple('Hello')
print(t6)
```

## Acessing Items
* Indexing
* Slicing

```python
print(t3)
print(t3[2])
print(t3[-4])
print(t4)
print(t4[:3])
print(t4[::-1])
print(t4[::3])
```

## Editing & Adding Items   

```python
t3[1] = 1000
# they are immutable
# adition is not possible in tuple
```

## Deleing items

```python
print(t3)
del t3
print(t3)
# they can delete all tuple elements but not works with single element deletion
```

## Operations on tuple

```python
# + and *
t1 = (1,2,3,4)
t2 = (5,6,7,8)

print(t1 + t2)
print(t1 * 3)
# membership

print(2 in t2)

# iteration
for i in t2:
    print(i)
```

## Tuple functions

```python
# len, max , min , sum , sorted
t = (34, 56, 90 ,-100)

print(len(t))
print(sum(t))
print(max(t))
print(min(t))
# ascending order
print(sorted(t))

# desecending order
print(sorted(t, reverse = True))
# count
t = (1,2,3,4,5,2,3,5,9,0,8,3)
# if not foun then return 0 in output
print(t.count(7))

print(t.count(3))
# index
print(t.index(2))
```

# **Set**
A set is a unoredered collection of items, Every set element is unique and must be immutable.

However, a set itself is mutable. we can add or remove elements.

Sets can alse perform mathematical set operations like union, intersection, symmetric difference stc.

## Charateristics
* Unordered
* Mutable
* No duplicate
* Cannot mutable data types
## Creating set

```python
# empty
s = set()
print(s)
print(type(s))

# 1D set
s1 = {1,2,3,4}
print(s1)

# homo and hetero
s2 = {1, 'hello', 3.14, True, (1,2,3)}   # true treated as a 1 that way only print 1 in output
print(s2)
# set not allow duplicates
s3 = {1, 2, 3, 4,4, 5, 4,6 }
print(s3)

# set cannnot have mutable items
s4 = {1,2,[3,4]}
print(s4)
```

## Adding items

```python
# add
s1 = {1,2,3,4}
s1.add(9)
print(s1)
 # update
s2 =  {1,2,3,4}
s2.update([5,6,7])
print(s2)
```

## Deletion in set

```python
# not allowed set element deletion
s1 = {1,2,3,4}
del s1
s2 = {1,2,3,4}
# discard: remoeve with value is entered
s2.discard(3)
print(s2)
# remove : if vaue not found thi sgive give an error 
s2.remove(1)
print(s2)
# pop: ya random hi koi element deletion krta hai 
s2.pop()
print(s2)
s2.clear()
print(s2)
```

## Set Operations

```python
a = {1,2,3,4,5}
b = {6,7,8,9,10, 4,5}
# union
print(a | b)
# intersection
print(a & b)
# difference
print(a - b)
# symmetric difference
print(a ^ b)
# membership test 
print(4 in a)
# loops in set
for i in a:
    print(i)
```

```python
# len , max, min , sorted, sum
s = {3,1,5,6,8,9}
print(sum(s))
print(len(s))
print(max(s))
print(min(s))
print(sorted(s))
# union
a = {1,2,3,4,5}
b = {6,7,8,9,10}
# same a | b
print('Union:',a.union(b))
# update: affects union result in set 1 
a.update(b)
print(a)
print(b)
a = {1,2,3,4,5}
b = {6,7,8,9,5,4,10}
# intersection same as a $ b
print('Intersection:',a.intersection(b))
# intersection_update
a.intersection_update(b)
print(a)
print(b)
# difference
a = {1,2,3,4,5}
b = {6,7,8,9,10,4,5}
# difference or difference_update
print('Difference:',a.difference(b))
a.difference_update(b)
print(a)
print(b)
```

```python
# syememtic difference
a = {1,2,3,4,5}
b = {6,7,8,9,10,4,5}
print('Symmetric Difference:',a.symmetric_difference(b))
a.symmetric_difference(b)
print(a)
print(b)
# isdisjoint/ issubset / issuperset
s1 = {1,2,3,4}
s2 = {3,4,5,6}
print('Disjoint:',s1.isdisjoint(s2))
s1 = {1,2,3,4,5}
s2 = {3,4,5}
print('Subset:',s2.issubset(s1))
print('Superset:',s2.issuperset(s2))
```

## Set comprehsnsion

```python
print({i for i in range(1,11)})
print({i * i for i in range(1,11) if i > 5})
```

# **Dictionary** 
**Dictionary** in python is a collection of these values used to store data values like a map which unlike other data types which holds only a single value has an element 

dict = {'name':'ali','age':23,'gender':'male'}

In some languages it is known as map or associative arrays
## Charactertistics
* Mutable
* Indexing has no meaning
* keys not duplicated
* keys cannot be mutable items

```python
# emppty dict 
d = {}
print(d)
# 1D dict
d1 = {'name':'ali','age':23,'gender':'male'}
print(d1)
# mixed keys
d2 = {(1,2,3):1,'age':'world'}
print(d2)
# 2D dict 
s = {
    'name' : 'Sami',
    'school': 'High',
     'sem': 4,
    'subjects': {
        'DSA': 50,
        'Math': 67,
        'English' : 78
    }
}
print(s)

# dict function
d4 = dict([(1,1),(2,2),(3,3),(4,4)])
print(d4)
```

## Acessing items

```python
d = {'name': 'ali', 'age': 23, 'gender': 'male'}
print(d)
# []
print(d['name'])
# get()
print(d.get('agi', 'N/A'))
```

## Adding Key-Value pairs

```python
d = {'name': 'ali', 'age': 23, 'gender': 'male'}
d['semester'] = 4
print(d)
d['percentage'] = 93.6
print(d)
```

## Removing key value pairs

```python
dic = {'name': 'ali', 'age': 23, 'gender': 'male', 'semester': 4, 'percentage': 93.6}
# pop
dic.pop('age')
print(dic)
# pop item
dic.popitem()
print(dic)
# del
del dic['name']
print(dic)
# clear 
dic.clear()
print(dic)
d2 = {'name': 'Sami', 'school': 'High', 'sem': 4, 'subjects': {'DSA': 50, 'Math': 67, 'English': 78}}
print(d2['subjects'])
print(d2['subjects']['Math'])
```

## Dictionary Operations
* Membership
* Iteration

```python
d2
print('Sami' in d2)
print('name' in d2)
d2 = {'name': 'Sami', 'school': 'High', 'sem': 4, 'subjects': {'DSA': 50, 'Math': 67, 'English': 78}}
for i in d2:
    print(i)
d2 = {'name': 'Sami', 'school': 'High', 'sem': 4}
for i in d2:
    print(i, d2[i])
```

## Dictionary Function

```python
len(d2)
print(d2)
print(sorted(d2))
# items/ keys/ values
print(d2.items())
print(d2.keys())
print(d2.values())
```

# **Dictionary Comprehension**

```python
# prinet first 10 numbers and thier square 
sq = {i : i * i for i in range(1, 11)}
print(sq)
# using exiting dict 
distances = {'Lahore' : 1000 , 'Karachi' : 2000 , 'Multan' : 3000}
dis = {key : value * 0.62 for (key , value) in distances.items()}
print(dis)# zip 
days = ['Mon','Tue','Wed','The','Thr','Fri','Sat','Sun']
temp = [32.8 , 34.6, 30.1 , 33.2, 29.8, 30.5, 31.8]
{i: j for (i, j) in zip(days, temp)}
```

```python
# nested table 
{i : {j : i * j for j in range(1, 11)} for i in range(2, 5)}
```

```python
# using if cond
products = {'Phone': 10 , 'Laptop' : 0 , 'Charger' : 32 , 'tablet' : 0}
{key : value for (key, value) in products.items() if value > 0}
```
